```python
"""
Created on 2020/6/13

@author: ShihYi,Yang
"""
#副程式

#新增
def insert(NBA):
    print('輸入球員名字 ex: Lebron James')
    print('')
    name = input("Enter a player's name:")
    print('')
    if name in NBA:
        print(name + " already exist in database.")
    else:
        alist = []
        team = input("the player's team:")
        score = float(input("生涯場均得分:"))
        rebound = float(input("生涯場均籃板:"))
        assist = float(input("生涯場均助攻:"))
        finals = int(input("總冠軍數:"))
        allstar = int(input("入選明星賽次數:"))
        alist.append(team)
        alist.append(score)
        alist.append(rebound)
        alist.append(assist)
        alist.append(finals)
        alist.append(allstar)
        NBA[name] = alist
        print('')
        print(name + ' 已加入資料中!')

#刪除
def delete(NBA):
    print('輸入球員名字 ex: Lebron James')
    print('')
    name = input("Enter a player's name:")
    print('')
    if name in NBA:
        YesorNo = int(input("Enter 1 if you want to delete the player's data, otherwise enter 0: "))
        if YesorNo == 1:
            del(NBA[name])
            print('')
            print("done.")
        else:
            print('')
            print("The player's data still exist.")
    else:
        print(name + " doesn't exist in database.")

#查詢
def search(NBA):
    newlist = ['所屬球隊:', '生涯場均得分:', '生涯場均籃板:', '生涯場均助攻:', '總冠軍數:', '入選明星賽次數:']
    print('輸入球員名字 ex: Lebron James')
    print('')
    name = input("Enter a player's name: ")
    print('')
    
    if name in NBA:
        for i in range(len(newlist)):
            print(newlist[i], NBA[name][i])
    else:
        print(name + " doesn't exist in database.")
        
#修改
def modify(NBA):
    print('輸入球員名字 ex: Lebron James')
    print('')
    name = input("Enter a player's name:")
    print('')
    if name in NBA:
        alist = []
        team = input("the player's team:")
        score = float(input("生涯場均得分:"))
        rebound = float(input("生涯場均籃板:"))
        assist = float(input("生涯場均助攻:"))
        finals = int(input("總冠軍數:"))
        allstar = int(input("入選明星賽次數:"))
        alist.append(team)
        alist.append(score)
        alist.append(rebound)
        alist.append(assist)
        alist.append(finals)
        alist.append(allstar)
        NBA[name] = alist
        print('')
        print("資料已修改!")
    else:
        print('')
        print(name + " doesn't exist in database.")

#列印出各項數據前十名
def top10(NBA):
    newDict = {}
    newlist = []
    i = int(input("輸入1:生涯場均得分, 輸入2:生涯場均籃板, 輸入3:生涯場均助攻, 輸入4:總冠軍數, 輸入5:入選明星賽次數     您的輸入:"))
    alist = ['所屬球隊:', '生涯場均得分:', '生涯場均籃板:', '生涯場均助攻:', '總冠軍數:', '入選明星賽次數:']
    
    for name in NBA:     #name = key
        x = NBA[name][i]     # x可能是得分、籃板.... , 看輸入的i為多少.
        if x not in newDict:
            newDict[x] = [name]
        else:
            newDict[x].append(name)
            
    x_list = list(newDict.keys())  # x_list裡面都是數據
    x_list.sort()
    x_list.reverse()
    
    print('')
    print(alist[i])    # ex: i=1, 生涯場均得分:
    for y in range(10):  #前10名
        for key in newDict:
            if x_list[y] == key:   # y從0~9, x_list[0]為最大值
                print(str(y+1) + '.', ", ".join( newDict[key] ), ':',  key)
                if x_list[y] == 0:
                    return None

#比較兩個球員之數據
def compare(NBA):
    print('輸入球員名字 ex: Lebron James')
    print('')
    name1 = input("Enter the first player's name:")
    if name1 not in NBA:
        print('')
        print(name1 + " doesn't exist in database.")
        return None
    name2 = input("Enter the second player's name:")
    if name2 not in NBA:
        print('')
        print(name2 + " doesn't exist in database.")
        return None
    n1_list = NBA[name1]     # ex: ['Los Angeles Lakers', 27.2, 7.4, 7.2, 3, 16]
    n2_list = NBA[name2]  # ex: ['Golden State Warriors', 23.5, 4.5, 6.6, 3, 6]
    newlist = ['所屬球隊', '生涯場均得分:', '生涯場均籃板:', '生涯場均助攻:', '總冠軍數:', '入選明星賽次數:']
    
    for i in range(1, len(newlist)):
        print('')
        print(newlist[i])
        if n1_list[i] > n2_list[i]:
            print(name1 + ' 勝!')
        elif n1_list[i] < n2_list[i]:
            print(name2 + ' 勝!')
        else:
            print("平手!")
```
